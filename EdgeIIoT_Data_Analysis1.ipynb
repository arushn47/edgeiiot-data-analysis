# ML-EdgeIIoT Dataset — EDA & Data Cleaning
**Dataset:** ML-EdgeIIoT-dataset.csv  
**Domain:** IoT Network Intrusion Detection  
**Task:** Multi-class attack classification from network traffic features

## 1. Imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', 50)
pd.set_option('display.float_format', '{:.4f}'.format)

plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False
```

## 3. Data Cleaning
### 3.1 Missing Values

```python
missing = data.isnull().sum()
missing_pct = (missing / len(data)) * 100
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing %', ascending=False)

if missing_df.empty:
    print("No missing values found in the dataset.")
else:
    print(missing_df)
    plt.figure(figsize=(12, 5))
    missing_df['Missing %'].plot(kind='bar', color='steelblue')
    plt.title('Missing Values (%)')
    plt.ylabel('Percentage')
    plt.tight_layout()
    plt.show()
```

### 3.2 Duplicate Rows

```python
dup_count = data.duplicated().sum()
print(f"Duplicate rows: {dup_count} ({dup_count/len(data)*100:.2f}%)")

data = data.drop_duplicates()
print(f"Shape after dropping duplicates: {data.shape}")
```

### 3.3 Drop Zero-Variance (Constant) Columns

```python
const_cols = [col for col in data.columns if data[col].nunique() <= 1]
print(f"Constant columns to drop: {const_cols}")

data = data.drop(columns=const_cols)
print(f"Shape after dropping constant columns: {data.shape}")
```

### 3.4 Fix Mixed-Type Columns

```python
# Columns flagged with mixed types — coerce to numeric where possible
mixed_type_cols = [
    'frame.time', 'arp.dst.proto_ipv4', 'arp.src.proto_ipv4',
    'http.file_data', 'http.request.uri.query', 'http.request.method',
    'http.referer', 'http.request.full_uri', 'http.request.version',
    'tcp.options', 'tcp.payload', 'tcp.srcport',
    'dns.qry.name.len', 'mqtt.conack.flags', 'mqtt.msg',
    'mqtt.protoname', 'mqtt.topic'
]

# Only process columns that still exist after previous drops
mixed_type_cols = [c for c in mixed_type_cols if c in data.columns]

converted, kept_as_str = [], []
for col in mixed_type_cols:
    coerced = pd.to_numeric(data[col], errors='coerce')
    non_null_orig = data[col].notna().sum()
    non_null_coerced = coerced.notna().sum()
    # If > 80% converts successfully, keep as numeric
    if non_null_orig == 0 or (non_null_coerced / max(non_null_orig, 1)) >= 0.8:
        data[col] = coerced
        converted.append(col)
    else:
        kept_as_str.append(col)

print(f"Converted to numeric: {converted}")
print(f"Kept as string (high cardinality / categorical): {kept_as_str}")
```

### 3.5 Cleaned Dataset Overview

```python
print("Final shape:", data.shape)
print("\nColumn type breakdown:")
print(data.dtypes.value_counts())
```

## 4. Exploratory Data Analysis
### 4.1 Target Variable — Attack Label & Attack Type

```python
print("Attack_label distribution:")
print(data['Attack_label'].value_counts())
print()
print("Attack_type distribution:")
print(data['Attack_type'].value_counts())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Binary label pie
label_counts = data['Attack_label'].value_counts()
axes[0].pie(
    label_counts,
    labels=['Attack (1)', 'Normal (0)'],
    autopct='%1.1f%%',
    colors=['#E24B4A', '#639922'],
    startangle=90
)
axes[0].set_title('Binary Class Distribution (Attack_label)')

# Attack type bar
type_counts = data['Attack_type'].value_counts()
type_counts.sort_values().plot(kind='barh', ax=axes[1], color='steelblue')
axes[1].set_title('Attack Type Distribution')
axes[1].set_xlabel('Count')

plt.tight_layout()
plt.show()
```

### 4.2 Protocol-Level Feature Analysis

```python
# Non-zero counts per protocol group
protocol_activity = {
    'ARP':       (data['arp.opcode'] > 0).sum(),
    'ICMP':      (data['icmp.checksum'] > 0).sum(),
    'HTTP':      (data['http.response'] > 0).sum(),
    'TCP':       (data['tcp.ack'] > 0).sum(),
    'UDP':       (data['udp.port'] > 0).sum(),
    'MQTT':      (data['mqtt.msgtype'] > 0).sum(),
}

proto_df = pd.Series(protocol_activity).sort_values(ascending=False)
proto_df.plot(kind='bar', color='teal', figsize=(10, 4))
plt.title('Active Rows per Protocol (non-zero count)')
plt.ylabel('Row Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### 4.3 Distribution of Key Numeric Features

```python
key_features = ['tcp.ack', 'tcp.len', 'tcp.seq', 'icmp.checksum',
                'http.content_length', 'mqtt.len', 'arp.opcode', 'udp.stream']
key_features = [f for f in key_features if f in data.columns]

fig, axes = plt.subplots(2, 4, figsize=(18, 8))
axes = axes.flatten()

for i, col in enumerate(key_features):
    vals = data[col].dropna()
    # Remove zeros to see non-trivial distribution
    vals_nonzero = vals[vals > 0]
    if len(vals_nonzero) > 10:
        axes[i].hist(vals_nonzero, bins=40, color='steelblue', edgecolor='white')
        axes[i].set_title(col, fontsize=10)
        axes[i].set_xlabel('Value')
        axes[i].set_ylabel('Frequency')
    else:
        axes[i].text(0.5, 0.5, 'All zeros', ha='center', va='center')
        axes[i].set_title(col, fontsize=10)

plt.suptitle('Distribution of Key Numeric Features (non-zero values)', fontsize=13, y=1.01)
plt.tight_layout()
plt.show()
```

### 4.4 Zero-Value Sparsity per Feature

```python
num_cols = data.select_dtypes(include='number').columns.drop(['Attack_label'], errors='ignore')
zero_pct = ((data[num_cols] == 0).sum() / len(data) * 100).sort_values(ascending=False)

plt.figure(figsize=(16, 5))
zero_pct.plot(kind='bar', color='coral')
plt.axhline(80, color='red', linestyle='--', linewidth=1, label='80% threshold')
plt.title('Zero-Value Sparsity per Numeric Feature (%)')
plt.ylabel('% Zero Values')
plt.legend()
plt.xticks(rotation=90, fontsize=8)
plt.tight_layout()
plt.show()

high_sparse = zero_pct[zero_pct >= 80]
print(f"Features with ≥80% zeros ({len(high_sparse)}):")
print(high_sparse.to_string())
```

### 4.5 Correlation Heatmap (Numeric Features)

```python
# Use a focused subset of non-sparse numeric features for readability
focus_cols = [
    'arp.opcode', 'arp.hw.size', 'icmp.checksum', 'icmp.seq_le',
    'http.content_length', 'http.response',
    'tcp.ack', 'tcp.checksum', 'tcp.connection.fin', 'tcp.connection.rst',
    'tcp.connection.syn', 'tcp.connection.synack', 'tcp.dstport',
    'tcp.flags', 'tcp.flags.ack', 'tcp.len', 'tcp.seq',
    'udp.stream', 'mqtt.len', 'mqtt.msgtype', 'Attack_label'
]
focus_cols = [c for c in focus_cols if c in data.columns]

corr = data[focus_cols].corr()

plt.figure(figsize=(16, 12))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(
    corr, mask=mask, annot=True, fmt='.2f',
    cmap='coolwarm', center=0,
    linewidths=0.4, annot_kws={'size': 7}
)
plt.title('Correlation Heatmap (Selected Numeric Features)', fontsize=13)
plt.tight_layout()
plt.show()
```

### 4.6 Feature Correlation with Target (Attack_label)

```python
corr_target = data[num_cols.tolist() + ['Attack_label']].corr()['Attack_label'].drop('Attack_label')
corr_target = corr_target.dropna().sort_values()

colors = ['#E24B4A' if v > 0 else '#378ADD' for v in corr_target]

plt.figure(figsize=(14, 6))
corr_target.plot(kind='bar', color=colors)
plt.axhline(0, color='black', linewidth=0.8)
plt.title('Feature Correlation with Attack_label')
plt.ylabel('Pearson r')
plt.xticks(rotation=90, fontsize=8)
plt.tight_layout()
plt.show()

print("Top positively correlated features:")
print(corr_target.tail(10).to_string())
print("\nTop negatively correlated features:")
print(corr_target.head(10).to_string())
```

### 4.7 Attack Type vs Key Features (Boxplots)

```python
box_features = ['tcp.len', 'icmp.checksum', 'http.content_length', 'arp.opcode']
box_features = [f for f in box_features if f in data.columns]

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes = axes.flatten()

for i, feat in enumerate(box_features):
    # Filter to non-zero for readability
    plot_data = data[data[feat] > 0][['Attack_type', feat]]
    order = plot_data.groupby('Attack_type')[feat].median().sort_values().index
    sns.boxplot(
        data=plot_data, x='Attack_type', y=feat,
        order=order, ax=axes[i],
        palette='Set2', showfliers=False
    )
    axes[i].set_title(f'{feat} by Attack_type (non-zero)', fontsize=10)
    axes[i].set_xlabel('')
    axes[i].tick_params(axis='x', rotation=45, labelsize=8)

plt.suptitle('Key Features by Attack Type', fontsize=13, y=1.01)
plt.tight_layout()
plt.show()
```

### 4.8 TCP Flag Distribution by Attack Type

```python
tcp_flags = ['tcp.connection.syn', 'tcp.connection.synack', 'tcp.connection.fin', 'tcp.connection.rst']
tcp_flags = [f for f in tcp_flags if f in data.columns]

flag_by_attack = data.groupby('Attack_type')[tcp_flags].sum()
flag_by_attack.plot(kind='bar', figsize=(14, 5), colormap='Set2', edgecolor='white')
plt.title('TCP Flag Counts by Attack Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(loc='upper right', fontsize=8)
plt.tight_layout()
plt.show()
```

### 4.9 Pairplot — Top Correlated Features

```python
top_features = abs(corr_target).nlargest(5).index.tolist()
pair_cols = top_features + ['Attack_label']
pair_cols = [c for c in pair_cols if c in data.columns]

# Sample for speed
sample = data[pair_cols].sample(n=3000, random_state=42)
sample['Attack_label'] = sample['Attack_label'].map({0: 'Normal', 1: 'Attack'})

sns.pairplot(sample, hue='Attack_label', palette={'Normal': '#639922', 'Attack': '#E24B4A'},
             plot_kws={'alpha': 0.3, 's': 10}, diag_kind='kde')
plt.suptitle('Pairplot — Top 5 Features Correlated with Attack_label', y=1.01)
plt.tight_layout()
plt.show()
```

### 4.10 Class Imbalance Summary

```python
attack_counts = data['Attack_type'].value_counts()

plt.figure(figsize=(12, 4))
attack_counts.plot(kind='bar', color='steelblue', edgecolor='white')
plt.axhline(attack_counts.mean(), color='red', linestyle='--', label=f'Mean: {attack_counts.mean():.0f}')
plt.title('Sample Count per Attack Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()

print("Imbalance ratio (max/min):", round(attack_counts.max() / attack_counts.min(), 2))
```
